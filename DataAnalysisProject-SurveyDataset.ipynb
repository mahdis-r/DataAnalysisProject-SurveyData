## Loading Data

```python
import pandas as pd
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/LargeData/m1_survey_data.csv"

filename = 'survey.csv'
urllib.request.urlretrieve(dataset_url,filename)

df = pd.read_csv(filename)

pd.set_option('display.max_columns', None)  # show all columns
df.head(3)
```

```python
df.shape
```

## Finding duplicates

```python
duplicated_rows = df[df.duplicated()]
len(duplicated_rows)
```

## Removing duplicates

```python
df.drop_duplicates(inplace=True)
print(df.shape)
```

## Finding Missing values

```python
for i in df.columns:
    print(str(i) + ": ", end='')
    print(df[str(i)].isna().sum())
```

```python
# doing a little heatmap
import seaborn as sns

corr_matrix = df.corr()

# Set figure size
plt.figure(figsize=(5, 5))

# Draw heatmap
sns.heatmap(corr_matrix,
            annot=True,      # show correlation values
            fmt='.2f',       # format to 2 decimals
            cmap='coolwarm', # color scheme
            cbar=True)       # show color bar

plt.title('Correlation Heatmap')
plt.show()
```

- So no two numerical features are really that correlated.

# Visualizations

### People surveyed

```python
# people surveyed 
counts = df['MainBranch'].value_counts()

# --- Pie Chart ---
plt.figure(figsize=(5, 5))
plt.pie(counts,
        labels=counts.index,       # category names
        autopct='%1.1f%%',         # show percentages
        startangle=90,             # start from top
        counterclock=False)        # clockwise order
plt.title('Distribution of Main Branch (Pie Chart)')
plt.show()
```

### Social Media used by employees

```python
print(f"{len(df['SocialMedia'].unique())} unique social media categories.\n")
print(f"The list includes:\n {df['SocialMedia'].unique()}\n")
print(df['SocialMedia'].value_counts())
```

```python
# Social Media used by employees
top_counts = df['SocialMedia'].value_counts()

plt.figure(figsize=(10, 5))
top_counts.plot(kind='barh', color='green', edgecolor='black')  # horizontal bar
plt.title('Social Media used by emplyees')
plt.xlabel('no. employees')
plt.ylabel('Social Media')
plt.gca().invert_yaxis()  # highest on top
plt.show()
```

```python
# Group, count, sort, and take top 5
df_top = df['DatabaseDesireNextYear'].value_counts().head(5)

# Extract data for the chart
data = df_top.values
labels = df_top.index

# Pie chart
plt.rcParams['figure.figsize'] = [5, 5]
fig, ax = plt.subplots()
ax.pie(data, labels=labels, autopct='%1.1f%%',
       explode=(0.1, 0, 0, 0, 0), shadow=True)
plt.title('Top 5 Database Desire Next Year')
plt.show()
```
